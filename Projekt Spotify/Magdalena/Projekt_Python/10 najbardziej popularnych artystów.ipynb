```python
import pandas as pd
import geopandas 
import fiona
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import folium
```

```python
spotify_database = pd.read_csv('Database to calculate popularity.csv')
```

```python
spotify_final_database = pd.read_csv('Final database.csv')
spotify_final_database
```

```python
top_global_songs = spotify_final_database[["Popularity", "Country", "Artist"]]
top_global_songs
```

```python
artist = top_global_songs.groupby(by = "Artist").sum().sort_values(by='Popularity', ascending = False).head(10)
artist
```

```python
artist.plot(kind = 'bar', color = ['green'], width = 0.7, ylabel = 'Popularity ', title = '10 najpopularniejszych artystów', grid =  True)
plt.grid(axis = 'x', which = 'major')
plt.tight_layout()
plt.savefig('wykres_artysci_2.png', dpi = 199)
```
