```python
import pandas as pd
import geopandas 
import fiona
import matplotlib as plt
import folium
```

```python
spotify_database = pd.read_csv('Database to calculate popularity.csv')
```

```python
spotify_final_database = pd.read_csv('Final database.csv')
```

```python
top_global_songs = spotify_final_database[["Popularity", "Country", "Title", "Artist"]]
top_global_songs
```

```python
top_global_songs.Title.value_counts().count() # ilość unikalnych piosenek w rankingu
```

```python
top_global_songs.Title.describe() # opis kolumny
```

```python
top_global_songs['rank'] = top_global_songs.groupby('Country')['Popularity'].rank(method = 'dense', ascending=False)
```

```python
top_global_songs
```

```python
top_glob = top_global_songs[(top_global_songs['Country'] == 'Global')].sort_values(by='Popularity', ascending = False).head(10)
top_glob
```

## 10 najbardziej popularnych piosenek na świecie != Global

```python
top_10_songs = top_global_songs[(top_global_songs['Country'] != 'Global')].sort_values(by=['rank', 'Country', 'Popularity'], ascending = True)
                                                                                                                                                
top_10_songs
```

```python
top_10_songs= top_10_songs[top_10_songs['rank'] <= 1.0]
```

```python
top_10_songs
```

```python
country_list = top_10_songs.Country.unique()
country_list
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
cities = geopandas.read_file(geopandas.datasets.get_path('naturalearth_cities'))
```

```python
world.at[143, 'name'] = 'UK'
world.at[143, 'name']
```

```python
world.at[4, 'name'] = "USA"
world.at[4, 'name']
```

```python
top_1_scal=world.merge(top_10_songs, how = 'left', left_on=['name'], right_on=['Country'])
```

```python
top_1_scal.name.unique()
```

```python
geo_top = top_1_scal.dropna(subset=['Country'])
```

```python
my_map = folium.Map()
choropleth = folium.Choropleth(
    geo_data=geo_top,
    name="top_10_songs",
    data=geo_top,
    columns=["Country", "Popularity", 'Title'],
    key_on="feature.properties.name",
    fill_color="OrRd",
    fill_opacity=0.8,
    line_opacity=0.4,
    legend_name="Popularity", 
).add_to(my_map)

folium.LayerControl(collapsed=True).add_to(my_map)
choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(        
        fields=['Country'
                ,'Title'
                ,'Artist'
                , 'Popularity'
               ],
        aliases=['Country: '
                ,'Title: '
                ,'Artist: '
                , 'Popularity: '
                 ],
        style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;") 
    )
)
my_map.save('songs_map.html')
```
