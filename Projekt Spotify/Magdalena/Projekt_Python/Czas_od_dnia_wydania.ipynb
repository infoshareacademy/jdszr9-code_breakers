```python
import pandas as pd
```

```python
spotify_database = pd.read_csv('Database to calculate popularity.csv')

```

```python
spotify_final_database = pd.read_csv('Final database.csv')
```

```python
 days_of_release = spotify_final_database[['Title', 'Artist', 'Country', 'Popularity', 'Days_since_release', "Release_date", "Released_after_2017"]]
days_of_release.head()
```

```python
days_of_release.sort_values(by='Days_since_release', ascending = True).head()
```

# Średni czas od dnia wydania do trafienia na listę. 
Uwaga !!!

Są tu uwzględnione piosenki które powstały na długo przed utworzeniem Spotify.

Średnia liczba dni, które upłynęły od czasu wydania utworu do trafienia na listę Spotify 200 wynosi: 1305 dni –  3 lata  6 miesięcy  i 27 dni. Jest to kryterium uwzględniające również  utwory, które zostały wydane na długo przed utworzeniem aplikacji Spotify oraz zestawienia Spotify top 200. 

```python
days_of_release.groupby('Title').mean('Days_since_release').mean() # Średni czas od dnia wydania piosenki 
```
